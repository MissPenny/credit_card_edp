EXPLORATORY DATA ANALYSIS ON CREDIT CARD TRANSACTIONS

The financial sector in the past years have been using Traditional fraud detection with rules based systems. The draw backs of the rule based system have been:
-Fixed thresholds per rule to determine fraud
-Limited to yes/no outcomes
-Fail to capture interaction between features

Machine learning algorithms are being developed to detect fraudulent transactions.In this project, we are going to explore a kaggle dataset containing information of transactions that occured over a period of two days to gain insight, find patterns and understand our data better.


```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

LOADING THE DATASET


```python
# loading the csv file using pandas
data = pd.read_csv('creditcard.csv')
```

```python
#exploring the dataset
print(data.columns)
```

```python
data.info
```

```python
#Sample data
data = data.sample(frac=0.05, random_state = 1)
print(data.shape)
print(data.describe())
```

```python
# V1 - V28 are the results of a PCA Dimensionality reduction to protect user identities and sensitive features
```

```python
# Plot histograms of each parameter 
data.hist(figsize = (20, 20))
plt.show()
```

```python
# Determine number of fraud cases in dataset

Fraud = data[data['Class'] == 1]
Valid = data[data['Class'] == 0]

outlier_fraction = len(Fraud)/float(len(Valid))
print(outlier_fraction)

print('Fraud transactions: {}'.format(len(data[data['Class'] == 1])))
print('Non Fraudulent transactions: {}'.format(len(data[data['Class'] == 0])))

```

```python
# Correlation matrix
corrval = data.corr()
visual = plt.figure(figsize = (12, 9))

sns.heatmap(corrval, vmax = 0.8, cmap='Blues_r')
plt.show()
```
